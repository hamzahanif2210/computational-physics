```python
# A Package/library is code written by someone which can be use by us (using import) to save our time. And it only need to be imported once. 
# For E.g: If want to calcualte a factorial we have to write a code or we can simply import from maths packages.

from math import factorial  # this statement states that import only factorial module(code) from math package.

print(factorial(22))


```

```python
#You can write your own code like below.

num = int(input("Enter a number: "))

factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,num + 1):
       factorial = factorial*i
   print("The factorial of",num,"is",factorial)
```

### Graph of sine and square 

```python
#another example

x = np.arange(-10*np.pi,10*np.pi,np.pi/100)
y = x**2
z = 1000*np.sin(x)
plt.plot(x,y, label='square of a')
plt.plot(x,z,label='sine of a')
plt.title("Two graphs in one")
plt.xlabel("Time")
plt.ylabel("Sqaure of a and Sine of a")
plt.legend(loc='best')
plt.grid(True)
plt.show() #plot graph
```
